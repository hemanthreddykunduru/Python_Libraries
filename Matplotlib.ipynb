# Bar-Graphs Plot_Matplotlib

## Overview-Syntax

```python
import matplotlib.pyplot as plt
x=["A","B","C","D"]
y=[1,2,3,4]
f = plt.figure()                                ## insted of plt.bar use plt.barh to make horizontal graph
f.set_figwidth(2)                               ## set_figwidth --adjust of x-axis length i.e width.
f.set_figheight(2)                              ## set_figheight --adjust of y-axis length i.e height.
plt.title("Title",fontsize=20)                  ## title --titleof the graph.
plt.xlabel('x - axis',fontsize=10)              ## xlabel --label of x-axis, fontsize --adjust in font size.
plt.ylabel('y - axis',fontsize=15)              ## ylabel --label of y-axis, fontsize --adjust in font size.
plt.bar(x,y,color="y",width=0.3,
        edgecolor="r",
        align="edge",
        linewidth=3,
        alpha=0.5)
## color --graph color, bar --bar graphs, width --width size of the bar,edgecolor --boundary color, align --aligning the bar's.
## linewidth --width of the edge line, alpha -- opacity
plt.show()
```

### adding diff color to the bar's

```python
c=["y","r","g","m"]
f = plt.figure()
f.set_figwidth(2)
f.set_figheight(2)
plt.xlabel('x - axis',fontsize=10)
plt.ylabel('y - axis',fontsize=15)
plt.bar(x,y,color=c, align="edge")  ## in align center maks us to locate bar's to center
plt.show()
```

## Labels

```python
c=["y","r","g","m"]
x=["A","B","C","D"]
y=[1,2,3,4]
plt.bar(x,y,color=c,label="yellow")
plt.legend()
plt.show()
```

# Stacked Bar_Graph

### coinsideing of graph 

```python
x=["A","B","C","D"]
y=[10,20,30,40]
z=[2,4,6,8]
plt.bar(x,y,color="y",label="men") ## combining both bars w.r.t to x-axis values
plt.bar(x,z,color="r",label="women")
plt.legend()                          ## legend --lableing of the data 
plt.show()
```

# Double Bar_Graph

```python
import matplotlib.pyplot as plt
import numpy as np
x=[2001,2002,2003,2004]
y=[10,20,30,40]
z=[2,4,6,8]

width=0.2
p=np.arange(len(x))  ## arrange is used to taking the position values
p1=width+p

plt.xlabel("Year's")
plt.ylabel("No.of")
plt.bar(p,y,width,color="y",label="men")
plt.bar(p1,z,width,color="r",label="women")
plt.xticks(p+width/2,x,rotation=90) ## rotation used to rotate the x axis values
plt.legend()
plt.show()
```

# Pie_Plot

```python
import matplotlib.pyplot as plt
x=[]
plt.pie(x)
plt.show()
```

```python
y=["A","B","C","D"]
x=[10,20,30,40]
exp=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=exp,autopct="%0.2f%%",
        shadow=True,radius=0.8,
       textprops={"fontsize":10}, wedgeprops={"linewidth":4})   ## explode --making of slice , autopct --diff w.r.t %
plt.legend(loc=2) ## loc --location of legend
plt.show()
```

```python
plt.pie(x,labels=y,
        shadow=True,radius=0.8,
       textprops={"fontsize":10},)
plt.pie([1],colors="w",radius=0.55)  ## overlapping of a dummy Pie_chart on a original chart
plt.show()
```
